**TOP 10 Merchant Revenue**

```python
dsrevenue = dataset.groupby(['merchant_name', 'merchant_area']).sum().sort_values(by = 'price', ascending = False)
dsrevenue.head(10)
```

**BOTTOM 10 Merchant Revenue**

```python
dsrevenue = dataset.groupby(['merchant_name', 'merchant_area']).sum().sort_values(by = 'price', ascending = True)
dsrevenue.head(10)
```

**TOP 10 Merchant Revenue in Surabaya**

```python
dst_1 = dataset[dataset['merchant_area'].str.contains('surabaya')]
dst_1 = dst_1.groupby(['merchant_name']).sum().sort_values(by = 'price', ascending = False)
dst_1.head(10)
```

**Bottom 10 Merchant Revenue in Surabaya**

```python
dst_2 = dataset[dataset['merchant_area'].str.contains('surabaya')]
dst_2 = dst_2.groupby(['merchant_name']).sum().sort_values(by = 'price', ascending = True)
dst_2.head(10)
```

**Top 10 Ordered Chinese Food in Gofood**

```python
dst_3 = dataset[dataset['product'].str.contains('Chinese')]
dst_3 = dst_3.groupby(['category']).count().sort_values(by = 'price', ascending = False)
dst_3.head(10)
```

```python
sns.barplot(
    x = "merchant_name",
    y = "product",
    data = dst_3
);

plt.title('Total Product Sales')
```
